#### 2.1 
Изучите распределение домов от наличия вида на набережную<br>
Постройте график<br>
Сделайте выводы

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('kc_house_data.csv', encoding='latin-1')

df.head()
```

```python
data = df['waterfront'].value_counts()
data.index = ['Нет', 'Есть']
plt.pie(data.values, labels=data.index, autopct='%1.2f%%')
plt.legend(data.index)
plt.title('Вид на набережную');
```

У большинства домов(99.25%) нет вида на набережную. У 0.75% - есть вид на набережную.
