#### 1.1. 
Скачать файл kc_house_data.csv - приложен на платформе<br>

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('kc_house_data.csv', encoding='latin-1')

df.head()
```

#### 1.2 
Изучите стоимости недвижимости

```python
plt.hist(df['price'])
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Стоимость недвижимости');
```

Большинство недвижимости стоит в пределах 700 000 - 800 000: более 17500 домов. На втором месте по количеству дома от 800 000 до 1 550 000: чуть меньше 2500шт. Немного домов до 3 900 000, дальше выбросы примерно до 8 000 000.

#### 1.3 
Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'])
plt.xlabel('Квадратура жилая')
plt.ylabel('Count')
plt.grid();
```

Большая часть домов - более 10000шт, квадратурой в пределах 1700 - 2800. На втором месте по квадратуре дома с 200 - 1700кв, около 7700шт. На третьем месте от 2800 до 4200, около 2800шт. Около 500шт в пределах 4200 - 5400. Мало домов от 5400 до 8200, дальше скорее выбросы до 14000.

#### 1.4 
Изучите распределение года постройки

```python
plt.hist(df['yr_built'])
plt.xlabel('Год постройки')
plt.ylabel('Count')
plt.grid();
```

Самое большое количество домов(более 3500), построенных в год, было с 2005 по 2015 год. С 1945 по 2005 год строилось примерно одинаковое количество домов в год(от 2600 до 3800), и за это время построилась большая часть домов в таблице. Намного меньше домов в год, строилось с 1922 по 1945 год(от 1200 до 1250), ещё немного меньше с 1900 по 1922 год(около 850).
