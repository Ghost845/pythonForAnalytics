```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
df
```

Новый столбец с название фирмы производителя процессора

```python
df['CPU_Company'] = df['Cpu'].apply(lambda x: x.split(' ')[0])
df.head()
```

Новый столбец только с гигабайтами, без типа хранилища

```python
def convert(x):
    x = x.split(' ')[0]
    if x.endswith('GB'):
        res = x.replace('GB', '')
    elif x.endswith('TB'):
        res = float(x.replace('TB', '')) * 1024
    return int(res)


df['Memory_Amount'] = df['Memory'].apply(convert)
df.head()
```

Новый столбец только с типом хранилища

```python
def convert2(x):
    x = x.split(' ')[1:]
    # x.pop[0]
    return ' '.join(x)
    # ' 'join(x.split(' ')[1:]) решение в одну строку


df['Memory_Type'] = df['Memory'].apply(convert2)
df.head()
```

Удаление двух столбцов

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace=True)
```

## EX2
Создать список SSD, со значениями = 0

```python
df['SSD'] = 0
df.head()
```

В список SSD изменить значение на 1, если у ноутбука есть SSD

```python
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1
df.head()
```

Поменять признак Weight, убрав kg и вернув вещественное значение

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float64')
```
